## Converting the miliseconds of duration column to seconds

```python
#using lambda function we can easily convert miliseconds to seconds

tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
```

```python
tracks.head()
```

As we can see, duration column has been populated at the end but duration_ms and duration both columns are still avalable in the dataframe. We need to drop column duration_ms

```python
#Dropping duration_ms column

tracks.drop("duration_ms", inplace = True, axis = 1)
tracks.head()
```

duration_ms has been dropped

# Plotting correlation map using heatmap from seaborn

```python
#dropping 3 columns in order to create the correlation map

df_corr = tracks.drop(["key","mode","explicit"], axis=1).corr(method = "pearson",numeric_only = True)
plt.figure(figsize=(15,6))
heatmap = sns.heatmap(df_corr, annot = True, fmt=".1g", vmin=-1, vmax = 1, center = 0, cmap = "inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

The correlation between energy and speechiness is negative correlation, that means when the energy is high speachiness tend to be low similary energy and loudness correlation is 0.8 that means when the enery is high loudness is also high.

As we all know that Acoustic music is often quiet and requires careful listening. That’s why it makes a negative correlation with energy and loudness, which makes sense.

# Regression Plots

```python
# Creating a sample dataframe using original dataframe in order to create a regression plot

sample_tracks = tracks.sample(int(0.004*len(tracks)))
print(len(sample_tracks))
```

0.4% of total dataset is 2346 rows

```python
# Plotting regression plot of loudness and Energy using regplot() of seaborn

plt.figure(figsize = (10,6))
sns.regplot(data = sample_tracks, y = 'loudness', x = 'energy', color = "c").set(title = "Loudness and Energy Regression plot")
```

As we can see the upward trend of regression line, there is strong positive correlation between loudness and energy. If the loudness is high, we can expect the energy to be high as well

```python
# Plotting regression plot of Acousticness and Loudness using regplot() of seaborn

plt.figure(figsize = (10,6))
sns.regplot(data = sample_tracks, y = 'loudness', x = 'speechiness', color = "c").set(title = "Speechiness and Loudness Regression plot")
```

As the loudness increases, the speechiness decreases

```python
# Plotting regression plot of Acousticness and Popularity using regplot() of seaborn

plt.figure(figsize = (10,6))
sns.regplot(data = sample_tracks, y = 'popularity', x = 'acousticness', color = "c").set(title = "Acousticness and Popularity Regression plot")
```

As the regression line is going downwards, that simply means as the acousticness increases, the popularity decreases

# Creating a year column with the help of release date column

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
#Checking if we have correct year column

years.head()
```

The populated years are correct in the year column

# Creating a Distribution plot to visualize the total number of songs in each year

```python
#Distribution plot of total number of songs in year

sns.displot(years, discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Number of songs per year")
```

According to distribution plot, we can say that since 1920, the number of songs released per year has increased in recent years.

# Plot for duration of song over the years

```python
#Creating plot for duration

total_duration = tracks.duration
fig, ax = plt.subplots(figsize = (18,7))
fig = sns.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title = "Year vs Duration Barplot")
plt.xticks(rotation = 90)
```

As we can see in above barplot, there is upward trend of duration from 1920 to 2010, But after 2010, the duration of songs started decreasing again.

# Line plot for average duration of the songs over the years

```python
#Line plot for average duration of the songs over the years

total_duration = tracks.duration
sns.set_style(style = "whitegrid")
fig, ax = plt.subplots(figsize = (18,6))
fig = sns.lineplot(x = years, y = total_duration, ax= ax).set(title = "Years vs Duration Line plot")
plt.xticks(rotation = 60)
```

As we can see in above lineplot, the duration of songs in 1920's was less. As the years passed by, the duration of songs increased till 2010. But after 2010, the duration of songs started decreasing again.
